On s'intéresse à la construction de la suite "look and say" ou "suite de Conway".

Cette suite part de $x_0=1$ et compte le nombre d'itérations à suivre de chaque élément.

Par exemple il y a un 1 (un 1) dans $x_0$ donc $x_1=11$.

Puis il y a deux 1  (deux 1) dans $x_1$ donc $x_2=21$

Il y a un 2 (un 2) et un 1 (un 1) dans $x_2$ donc $x_3=1211$

Il y a un 1 (un 1) puis un 2 (un 2) puis deux 1 (deux 1) donc $x_4=111221$

<div class="alert alert-danger" role="alert">
Pour simplifier le problème chaque $x_i$ est stocké dans une liste.
    
Par exemple $x_1$=`[1,1]` et $x_4$= `[1,1,1,2,2,1]`
</div>
