<button type="button" class="btn btn-info btn-lg btn-block"><h4><div align="left"><i>Exercice 4</i></div></h4></button>

<div class="alert alert-danger" role="alert">
  Écrire un programme qui calcule les racines carrées de nombres saisis au clavier.
</div>

<hr> 

Ce programme s’arrêtera lorsqu’on lui fournira la valeur 0. Il refusera les valeurs négatives. Afin d’obtenir la racine carrée d’un nombre, on pourra utiliser la méthode `sqrt()` du module `math` :

```python
import math
math.sqrt(2)
```

Exemple de résultat attendu :

<div class="alert alert-warning" role="alert">
  Votre code :
</div>

```python
import math
a = True
while a:
    a = float(input("Veuillez entrez un nombre positif : "))
    if a < 0:
        print('Ce nombre est négatif !')
    elif a > 0:
        print(math.sqrt(a))
    else:
        print('bye!')
```

<button type="button" class="btn btn-info btn-lg btn-block"><h4><div align="left"><i>Exercice 5 (difficile)</i></div></h4></button>

<div class="alert alert-danger" role="alert">
  Écrire un programme qui affiche un triangle isocèle formé d’étoiles.
</div>

<hr> 

La hauteur du triangle (c’est-à-dire son nombre de lignes) sera saisie au clavier. On s’arrangera pour que la dernière ligne du triangle s’affiche sur le bord gauche de l’écran.

<hr> 

Exemple d'affichage :

<div class="alert alert-warning" role="alert">
  Votre code :
</div>

```python
nb_ligne = int(input('Combien de ligne(s) à afficher ? '))
print()
for l in range(1, nb_ligne + 1):
    nb_espace = nb_ligne - l
    nb_etoile = 2 * l - 1
    print(' ' * nb_espace + '*' * nb_etoile)
```

<button type="button" class="btn btn-info btn-lg btn-block"><h4><div align="left"><i>Exercice 6 (difficile)</i></div></h4></button>

<div class="alert alert-danger" role="alert">
  Écrire un programme qui affiche toutes les manières possibles d'obtenir un euro avec des pièces de 1 centime, 2 centimes, 5 centimes et 10 centimes.
</div>

<hr> 

Cet exercice est un calcul de combinaisons. Les résultats devront être présentés ainsi (il s’agit d’un extrait !) :

<div class="alert alert-warning" role="alert">
  Votre code :
</div>

```python
for un in range(101):
    for deux in range(51):
        for cinq in range(21):
            for dix in range(11):
                if un + deux * 2 + cinq * 5 + dix * 10 == 100:
                    print(un, 'pièce(s) de 1,', deux, 'pièce(s) de 2,', cinq, 'pièce(s) de 5,', dix, 'pièce(s) de 10.')
```

<button type="button" class="btn btn-info btn-lg btn-block"><h4><div align="left"><i>Exercice 7 (difficile)</i></div></h4></button>

<div class="alert alert-danger" role="alert">
  Ecrivez un programme qui permet d'afficher les nombres premiers compris entre 1 et 100.
</div>

<div class="alert alert-warning" role="alert">
  Votre code :
</div>

```python
for i in range(2, 101):
    prime = True
    j = 2
    while prime and j <= i // 2:
        prime = i % j != 0
        j += 1
    if prime:
        print(i) 
```
