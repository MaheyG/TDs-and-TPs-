<div class="alert alert-danger" role="alert">
On chercher une approximation de $\pi$, pour cela on s'intéresse maintenant à la suite:
   $$x_n=\frac{-1^n}{2n+1}$$
    
Attention la suite part de $n=0$
    
    
Remarquez que $-1^n=1$ quand $n$ est paire et $-1^n=-1$ quand $n$ est impaire.

<div class="alert alert-warning" role="alert">
Soit n un paramètre, ecrire une boucle qui donne la somme suivante
    $$x_0+x_1+...+x_{n-1}+x_{n}$$

<div class="alert alert-warning" role="alert">
La quantité $4\times (x_1+...+x_n)$ tend vers $\pi$. Soit k un paramètre, ecrire une boucle ```while``` qui approxime la k-ième décimal de $\pi$.
</div>

````python
import math
# pi est donné par :
pi = math.pi
print(pi)

# L'approximation à la k-ième décimal est donnée par :
k=3
pi_approx = round(pi*10**k)/10**k
print(pi_approx)
````
