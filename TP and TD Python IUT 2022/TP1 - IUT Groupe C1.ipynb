<div class="alert alert-danger" role="alert">
On chercher une approximation de $\pi$, pour cela on s'intéresse maintenant à la suite:
   $$x_n=\frac{-1^n}{2n+1}$$
    
Attention la suite part de $n=0$
    
    
Remarquez que $-1^n=1$ quand $n$ est paire et $-1^n=-1$ quand $n$ est impaire.
    
   
    
Ecrire une fonction qui retourne les n  premiers termes de la suite $(x_n)$ dans une liste
</div>

```python
def x_n(p):
    result=[]
    for i in range(p):
        x_n=(-1)**i/(2*i+1)
        result=result+[x_n]
    return result
print(x_n(10))
```

<div class="alert alert-warning" role="alert">
  Retournez dans une 2nd liste la valeur de la somme 
    $$x_0+x_1+...+x_{n-1}+x_{n}$$ 
    à chaque itérations. 
</div>

```python
def somme_x_n(p):
    result=[]
    somme=[0]
    for i in range(p):
        x_n=(-1)**i/(2*i+1)
        result=result+[x_n]
        somme=somme+[somme[-1]+x_n]
    return result,somme
_,b=somme_x_n(10)
print(b)
```

<div class="alert alert-warning" role="alert">
La quantité $4\times (x_1+...+x_n)$ tend vers $\pi$. Ecrire une fonction approx(k) qui renvoie le nombre d'itérations de la suite $4\times(x_n)$ nécéssaire pour approximer la k-ième décimal de $\pi$.
</div>

```python
import math
# pi est donné par :
pi = math.pi
print(pi)

# L'approximation à la k-ième décimal est donnée par :
k=3
pi_approx = round(pi*10**k)/10**k
print(pi_approx)
```

<div class="alert alert-warning" role="alert">
Par exemple la première décimal de $\pi$ est 3.1 qui est approximé par $x_0+x_1+...+x_{17}$. La fonction doit renvoyer 17.
</div>

```python
def approx(k):
    import math
    pi = math.pi
    pi_approx = round(pi*10**k)/10**k
    i=1
    _,somme_xn=somme_x_n(i)
    b=4*somme_xn[-1]
    print(b)
    x_n_approx=round(b*10**k)/10**k
    while x_n_approx != pi_approx :
        i+=1
        _,somme_xn=somme_x_n(i)
        b=4*somme_xn[-1]
        x_n_approx=round(b*10**k)/10**k
        print(x_n_approx)
    return i
```
