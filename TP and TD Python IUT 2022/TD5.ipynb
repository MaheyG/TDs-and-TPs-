<div class="alert alert-danger" role="alert">
  Écrire une fonction qui permettra d'afficher un diagramme en barre à partir de données passées en arguments.
</div>

Les arguments devront être deux tuples (ou listes) : `noms` et `valeurs`:
* `noms` devra contenir les étiquettes des barres
* `valeurs` contiendra les valeurs

Vous devrez vérifier que :
* les deux tuples (ou listes) ont la **même taille**
* les tuples (ou listes) n'ont **pas plus de 10 éléments**
* `noms` ne contient que des **chaines des caractères**
* `valeurs` ne contient que des **nombres flottants**

Exemple d'affichage de diagramme en barre en python :

```python
%matplotlib inline
import matplotlib.pyplot as plt

labels = ('a', 'b', 'c', 'd')
values = (10, 8, 11, 4)

plt.bar(labels, values) ; 
plt.show()
```
