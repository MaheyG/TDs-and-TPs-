**Rappels**

* En Python, chaque variable a une <span style="color:red">identité</span>, un <span style="color:red">type</span> et une <span style="color:red">valeur</span>

Les types "séquences" en python sont (nous reviendrons dessus lors du TD2) :
* les chaines de caractères : `str`
* les listes : `list`
* les tuples : `tuple`

Pour connaitre le type d'une variable `var` : `type(var)`

<button type="button" class="btn btn-info btn-lg btn-block"><h4><div align="left"><i>Exercice 4 : type et opérateur</i></div></h4></button>

**Rappels**

Une <span style="color:red">expression</span> est un calcul faisant intervenir une ou plusieurs valeurs (les <span style="color:red">opérandes</span>) pour une opération (via des <span style="color:red">opérateurs</span>) déterminée.

La <span style="color:red">dimension</span> d'un opérateur est déterminer par son nombre d'opérandes. Un opérateur <span style="color:red">unaire</span> possède une opérande, etc.

*Opérateurs numériques (non exhaustif)*
* opérateurs unaires :
    * `-` (négatif), 
    * `+` (positif)
* opérateurs binaires : 
    * `/` (division), 
    * `//` (division entière),
    * `%` (modulo),
    * `*` (multiplication), 
    * `**` (puissance), 
    * `+` (addition), 
    * `-` (soustraction)
    
*Opérateurs de chaines de caractères binaires (non exhaustif)*
* `+` (concaténation)
* `*` (duplication)
    
*Opérateurs de comparaison*
* `<` (inférieur)
* `>` (supérieur)
* `==` (égal)
* `>=` (sup. ou égal)
* `<=` (inf. ou égal)
* `!=` (différent)
* `in` (inclusion)

*Priorité des opérateurs (non exhaustif)*


<div class="alert alert-danger" role="alert">
  Pourquoi peut-on écrire ceci :
</div>

```python
a = 1 == 3 + 2
```

<div class="alert alert-warning" role="alert">
  Votre explication :
</div>

Car : 
* 1 == 3 est un booléen qui, en python, est considéré comme un entier si utilisé dans une expression numérique
* les opérateurs de calcul sont prioritaires sur les opérateurs de test en python

Ainsi, ici, on a :
* 3 + 2 qui vaut 5
* 1 == 5 qui vaut False
* a qui vaut finalement False

**Rappels**

En Python, les <span style="color:red">blocs d'instruction</span> sont identifiés par l'<span style="color:red">indentation</span>.

Structure logique du `if` :

Exemple :

```python
a = 6
if a == 0:
    print("valeur nulle")
elif a > 0:
    print("valeur positive")
else:
    print("valeur négative")
```

Vous pouvez notamment lire la documentation de `input` (https://docs.python.org/fr/3/library/functions.html#input) pour vous aider.

<div class="alert alert-warning" role="alert">
  Votre explication :
</div>

  Votre code corrigé :

```python
a = input('saisissez un nombre : ')
if a == '2':
    print('vous gagnez !')
else:
    print('vous perdez')
```

<div class="alert alert-danger" role="alert">
  Modifiez votre code précédent afin de vérifier que l'utilisateur saisit bien un entier.

<hr>

Vous pourrez utiliser pour cela la méthode `isdigit()`

<div class="alert alert-warning" role="alert">

<button type="button" class="btn btn-info btn-lg btn-block"><h4><div align="left"><i>Exercice 8 : boucle for (1/2)</i></div></h4></button>

**Rappels**

L'instruction `for` est utilisée pour <span style="color:red">itérer</span> sur les éléments d'une séquence (par exemple une chaîne de caractères, un tuple, une liste, etc.).

Itérer : fait de répéter une action en parcourant un objet itérable (comme une séquence en Python).

Une instruction `break` exécutée dans la première "suite" termine la boucle sans exécuter la "suite" de la clause `else`. 

Une instruction `continue` exécutée dans la première "suite" saute le reste de la "suite" et continue avec l'élément suivant (ou avec la clause `else` s'il n'y a pas d'élément suivant)

Structure logique du `for` :

Exemple (ici nous itérons sur la liste [1, 2, 3]) :

```python
somme = 0
for a in [1, 2, 3]:
    somme += a
print(somme)
```

Exemple d'instruction `for` avec une itération sur une liste d'entiers contigus (`range(3)`) :

```python
for a in range(3):
    print(a)
```
